```python
merged = map_data.set_index('Name').join(df.set_index('States'))#.reset_index(level=0, inplace=True)
merged#.head()
```

```python
import matplotlib.pyplot as plt
import geopandas as gpd

# Assuming 'merged' is a GeoDataFrame with a 'Persons' column for plotting
fig = plt.figure(figsize=(12, 8), dpi=1200)

# Create the plot using the 'RdPu' colormap, disable the GeoDataFrame legend
ax = merged.plot(column='Birth_reg', categorical=False, legend=False,
                  cmap='Blues', edgecolor='black', linewidth=0.25)

# Customize the colorbar legend with reduced length (shrink=0.7)
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=merged['Birth_reg'].min(), vmax=merged['Birth_reg'].max()))
sm._A = []
cbar = plt.colorbar(sm, ax=ax, pad=0.02, shrink=0.7)  # Adjust the shrink parameter as needed

# Remove x-axis and y-axis ticks and frame
plt.xticks([])
plt.yticks([])
ax.set_frame_on(False)

# Save the plot
plt.savefig('/home/debodeep94/OneDrive/Indian researcher/desidata/nfhs/nfhs.svg', bbox_inches='tight')

# Show the plot (optional)
plt.show()
```
